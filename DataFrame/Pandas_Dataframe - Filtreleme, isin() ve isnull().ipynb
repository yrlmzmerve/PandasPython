# DataFrame - Filtreleme, isin() ve isnull() 

```python
import pandas as pd
```

```python
milli = pd.read_csv('millitakım.csv')
milli.head()
```

## Filtreleme

```python
#Takımları Görme
milli["Takım"].head(2)
```

```python
#Fenerbahçe olup olmamasına göre True False değeri döndürür :
# milli["Takım"]=="Fenerbahçe"
takım = milli["Takım"]=="Fenerbahçe"
```

```python
#liste içinde Fenerbahçe olanları getirme
milli[takım]
```

```python
# Görevi önlibero olanarı filtreleme
onlibero = milli["Görev"] == "önlibero"
milli[onlibero]
```

```python
#takım = Fenerbahçe, görev=önlibero olanları getirme
milli[takım & onlibero]
```

```python
# takım = Bursa , gçrev = kaleci
bursa = milli["Takım"] == "Bursa"
gor = milli["Görev"] == "kaleci"
milli[bursa & gor]
```

```python
#yaşı 25'ten büyük olan, son 3 sporcu
yas = milli["Yaş"]>25
milli[yas].tail(3)
```

```python
#yaşı 25ten büyük ya da görevi önlibero olanları getirme
milli[yas | onlibero].head(4)
```

```python
#yaşı 25'ten büyük ve görevi =önlibero ya da takımı=fb olanlar
milli[yas & onlibero | takım]
```

### isin() func.

```python
# listede var mı yok mu bakmak için
# milli["Takım"].isin(["Fenerbahce","Kayseri"])
# yoksa false , varsa true dondurur
fenkay = milli["Takım"].isin(["Fenerbahçe","Kayseri"])
```

```python
milli[fenkay]
```

### isnull()

```python
milli["Gol"].head(3)
```

```python
# Nan değermi diye bakar ; true false dondurur
# milli["Gol"].isnull()
null = milli["Gol"].isnull()
```

```python
# Gol sütunu nan olanları getirme
milli[null].head(3)
```

### notnull()

```python
# Gol sütunu notnan olanları getirme
notnull = milli["Gol"].notnull()
milli[notnull].head(3)
```
