# DataFrame - Birden fazla index

```python
import pandas as pd
```

```python
world = pd.read_csv('wcplayers.csv')
world.head(3)
```

### set_index()

```python
#Team - Shirt Name 'i  en baştaki sütunlara getirme
world.set_index(keys=["Team","Shirt Name"], inplace=True)
```

```python
#Alfabeye göre sıralama (Team ve Shirt Name'de)
world.sort_index().head(3)
```

```python
#indexin isimleri
world.index.names
```

```python
world.index[0]
```

### get_level_values(index) veriyi çekme

```python
world.index.get_level_values(0)
```

```python
world.index.get_level_values(1)
```

```python
# index isimlerini değiştirme 
world.index.set_names(["Takım","Takma İsim"], inplace=True)
```

```python
world.head(1)
```

```python
# Sıralama da 
# Takımlar b->k , Takma isimler k->b sıralanacak
world.sort_index(ascending=[False,True]).head(3)
```

```python
# indexlerin erleri değiştirmek için: Takma İsim ile Takım 
world.swaplevel().head(3)
```

```python
#world.stack()

#frame şeklinde listelemek için
a = world.stack().to_frame().head()
a
```

```python
# Stack işlemini geri almak için
a.unstack().head()
```

```python
# pivot
futbol = pd.read_csv('pivot.csv')
futbol
```

```python
futbol.pivot(index="Futbolcu", columns="Mevsim", values="Gol")
```

```python
# toplam golleri
futbol.pivot_table(values="Gol", index="Futbolcu", aggfunc="sum")
```

```python
futbol.pivot_table(values="Gol", index="Futbolcu", aggfunc="mean")
```

```python
# melt 

m = pd.read_csv('melt.csv')
m
```

```python
pd.melt(m,id_vars="Futbolcu", var_name="Yıl" , value_name="Gol")
```
