```python
takım = oyun.groupby("Team")
```

```python
#Dataframe oluşturma
df = pd.DataFrame(columns=oyun.columns)
```

```python
# oluşturulan df'in for ile içini doldurma
for ülke,data in takım:
    uzun=data.nlargest(1,"Height")
    df=df.append(uzun)
```
