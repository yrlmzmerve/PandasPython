# Dataframe between{}, dublicated{}, drop_dublicates{}, unique{}

```python
import pandas as pd
```

```python
dataframe = pd.read_csv('millitakım.csv')
dataframe.head(0)
```

```python
dataframe.info()
```

```python
dataframe["Milyon Euro"] = dataframe["Milyon Euro"].astype("float")
dataframe.head(1)
```

### between

```python
# 20 -21 arasını alma (21de dahil)
yas = dataframe["Yaş"].between(20,21)
dataframe[yas]
```

```python
milyon = dataframe['Milyon Euro'].between(5,10)
dataframe[milyon]
```

```python
# numarası 5-8 arası olanlar (8 dahil)
numara = dataframe["Numara"].between(5,8)
dataframe[numara]
```

### dublicated

```python
# dublicate olanlarda ilkini siler
# örneğin 3 tane dul var 1. siler : keep = first oldugunda.
# keep=last olursa en sondakini siler
df = dataframe["Görev"].duplicated(keep='last')
dataframe[df].head()
```

### drop_dublicated()

```python
#dublicateleri siler sadece ilki kalır
dataframe.drop_duplicates(subset="Görev")
```

### unique()

```python
dataframe["Görev"].unique()
```

```python
#kaç tane değerin geçtiğini bulmak için
len(dataframe["Görev"].unique())
```
