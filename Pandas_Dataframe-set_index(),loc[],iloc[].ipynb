# Dataframe set_index(), loc[], iloc[]

```python
import pandas as pd
```

```python
df = pd.read_csv('millitakım.csv')
df.head(1)
```

### set_index()

```python
# Liste basında gösterilen indexler (0,1,.) çıkıp ilk sütun isim oldu
# inplace = True ile kalıcılastırdık

df.set_index("İsim",inplace=True)
df.head(3)
```

```python
# set_index ile yapılan değişikliği reset_index ile geri alma
# liste başındaki indexler geri gelir

df.reset_index(inplace=True)
df.head(3)
```

### loc []

```python
#değere göre veriyi verme
df = pd.read_csv('millitakım.csv',index_col="Numara")
df.head(3)
```

```python
# Numara değeri 12 olan veriyi getirme
df.loc[12]
```

```python
df.set_index("İsim",inplace=True)
```

```python
df.head(6)
```

```python
# sinan ile kaan arasındaki oyuncuları getirir kaan dahil
df.loc["Sinan Bolat":"Kaan Ayhan"]
```

### iloc []

```python
#indexlere göre veriyi getirme
df.iloc[5]
```

```python
# 0'dan 3e kadar olanları alma
df.iloc[0:3]
```

```python
#0.indexteki verinin 3. verisini alma
df.iloc[0,3]
```
