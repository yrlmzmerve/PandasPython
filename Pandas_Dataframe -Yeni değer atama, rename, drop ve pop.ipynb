# Dataframe yeni değer atama, rename(), drop() ve pop() 

```python
import pandas as pd
```

```python
df = pd.read_csv('millitakım.csv', index_col="İsim")
df.head()
```

```python
df.iloc[4,0]
```

### Değer Atama

```python
#numaralarını değiştirme
df.iloc[[4,5,6],0]
```

```python
df.iloc[[4,5,6],0] = [100,200,300]
df.iloc[[4,5,6],0]
```

```python
#Takımı fenerbahçe olanları al
#Takımlarını değiştir
f = df["Takım"] == "Fenerbahçe"
```

```python
df.loc[f,"Takım"] = "Galatasaray"
df[f]
```

### rename()

```python
#Sütun adını değiştirme -> Görev - Mevki , Maç - Maç Sayısı
df.rename(columns= {"Görev":"Mevki", "Maç": "Maç Sayısı"})
df.head(3)
```

```python
df.rename(index= {"Harun Tekin":"HARUN TEKİN"})
df.head(1)
```

### drop methodu
##### Satır ve sütunu silmek için

```python
#axis=0 satırda değşiklik, axis=1 sütunda değişiklik

df.drop(labels=["Maç","Gol"], axis=1)
df.head(3)
```

### Pop metodu

```python
#Silinen bölümü verir ve df'den Yaş Sütununu siler
df.pop("Boy").head(3)
```
