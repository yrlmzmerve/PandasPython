# DataFrame yeni sütun ekleme, çıkarma, güncelleme

```python
dunya = pd.read_csv('dünya.csv')
dunya
```

```python
dunya.head()
```

```python
# Ülke sütununu alma
dunya[["Ülke"]].head()
```

```python
dunya[["Ülke","KITA"]].head()
```

```python
#Yeni sütun ekleme
dunya["Sıralama"] = "top 20"
dunya
```

```python
# İki sütun arasına eleman eklemek için Insert kullanılır
# 2.sütuna ekle
dunya.insert(2, column="CN", value="DN")
```

```python
dunya
```

```python
#Sütun değerlerini değiştirme
dunya["Gelir"] = dunya["Gelir"]/100000
```

```python
dunya
```

```python
#Yeni sütun oluşturulu, mevcut Gelir sutonundi degerler 5.82 ile carpılark eklenen sütuna yazılır
dunya["Gelir TL"] = dunya["Gelir"]*5.82
```
